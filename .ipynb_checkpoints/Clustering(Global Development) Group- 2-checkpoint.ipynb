```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
#Load the Dataset
data = pd.read_excel('World_development_mesurement.xlsx')
data.head()
```

```python
df = data.copy()
```

```python
df.dtypes
```

```python
#Converting the object to float
monetary_columns = ['GDP', 'Health Exp/Capita', 'Tourism Inbound', 'Tourism Outbound']

# Remove non-numeric characters ('$', ',') and convert to numeric
for col in monetary_columns:
    df[col] = df[col].replace('[\$,]', '', regex=True).astype(float)

# Check if the conversion was successful
df[monetary_columns].head()
```

```python
df['Business Tax Rate']=df['Business Tax Rate'].astype(str).str.replace('%','',regex=True)
df['Business Tax Rate'] = pd.to_numeric(df['Business Tax Rate'], errors='coerce')
```

```python
# This step creates a summary of data where each row represents the average metrics for a specific country
grouped_df = df.groupby('Country', as_index=False).mean()

# Drop rows with missing values (NaN) to ensure clean and reliable data for further analysis
grouped_df = grouped_df.dropna()

# Display the resulting grouped and cleaned DataFrame
grouped_df
```

```python
# Get the top 10 countries with the highest GDP
top_10_countries = grouped_df.sort_values('GDP', ascending=False).head(10)

# Plot a bar chart for the top 10 countries by GDP
plt.figure(figsize=(10, 6))
plt.bar(top_10_countries['Country'], top_10_countries['GDP'], color='skyblue')

# Add labels and title to the chart
plt.xlabel('Country')
plt.ylabel('GDP (in billions)')
plt.title('Top 10 Countries by GDP')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Adjust layout and display the chart
plt.tight_layout()
plt.show()
```

- United States have a highest average GDP among other countries

```python
# Sort by Population Total and get the top 10 countries
top_10_population = grouped_df.sort_values('Population Total', ascending=False).head(10)

# Plot the top 10 countries by population as a bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_10_population['Country'], top_10_population['Population Total'], color='orange')

# Add labels and title
plt.xlabel('Country', fontsize=12)
plt.ylabel('Population (in billions)', fontsize=12)
plt.title('Top 10 Countries by Population', fontsize=14)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

## Distribution Plot 

```python
# Get the column names of the DataFrame
columns = df.columns

# Create a 6x4 grid of subplots for visualizing distributions
fig, axes = plt.subplots(6, 4, figsize=(14, 12), sharex=False, sharey=False)

# Flatten the axes array for easier iteration
axes = axes.flatten()

# Loop through each column and plot its distribution
for i, col in enumerate(columns):
    if i < len(axes):  # Ensure we don't exceed the available axes
        # Plot the distribution of the column, drop missing values, and add KDE
        sns.distplot(df[col].dropna(), ax=axes[i], kde=True)
        axes[i].set_title(col)  # Set the title of the subplot to the column name
    else:
        break  # Stop if there are more columns than available subplots

# Adjust layout to prevent overlap of subplots and labels
plt.tight_layout(pad=2.0)

# Show the plot
plt.show()
```

```python
# Lists to track updated columns with their skewness for future reference
mean_updated_columns = []
median_updated_columns = []

# Loop through columns with numeric types (float64 and int64)
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    if df[col].isna().sum() > 0:  # Check if the column contains any missing values
        skewness = df[col].skew()  # Calculate skewness of the column
        
        # Check if the skewness is below the threshold (indicating near-normal distribution)
        if abs(skewness) < 0.5:  # Assumed threshold for normal-like distribution
            # Replace missing values with the mean of the column
            df[col] = df[col].fillna(df[col].mean())
            mean_updated_columns.append((col, skewness))  # Track column and its skewness
        else:
            # Replace missing values with the median of the column
            df[col] = df[col].fillna(df[col].median())
            median_updated_columns.append((col, skewness))  # Track column and its skewness

# Print the list of columns updated with the mean and their skewness values
print("Columns updated with mean (skewness):")
for col, skewness in mean_updated_columns:
    print(f"{col}: {skewness:.4f}")

# Print the list of columns updated with the median and their skewness values
print("\nColumns updated with median (skewness):")
for col, skewness in median_updated_columns:
    print(f"{col}: {skewness:.4f}")
```

## **Observations:**
1. Columns like **GDP, CO2 Emissions, and Lending Interest** have extremely high skewness, indicating significant outliers or uneven distributions.
2. **Life Expectancy Female and Life Expectancy Male** are negatively skewed but still not close enough to normal, so median imputation was used.

```python
# Loop through each column in the DataFrame
for col in df.columns:
    # Calculate the first quartile (25th percentile) and third quartile (75th percentile)
    Q1 = df[col].quantile(0.25)  
    Q3 = df[col].quantile(0.75) 
    
    # Compute the Interquartile Range (IQR)
    IQR = Q3 - Q1              
    
    # Define lower and upper bounds for detecting outliers
    lower_bound = Q1 - 1.5 * IQR  
    upper_bound = Q3 + 1.5 * IQR  
    
    # Cap outliers by replacing values below the lower bound with the lower bound
    # and values above the upper bound with the upper bound
    df[col] = np.where(df[col] < lower_bound, lower_bound, 
                       np.where(df[col] > upper_bound, upper_bound, df[col]))
```

### K-Means CLustering 

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Elbow Method
sse = []
range_n_clusters = range(2, 11)
for k in range_n_clusters:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_df)
    sse.append(kmeans.inertia_)

plt.plot(range_n_clusters, sse, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('SSE (Sum of Squared Errors)')
plt.title('Elbow Method')
plt.savefig('elbow.jpeg')
plt.show()
```

### Elbow Method Analysis
1. The "elbow" in the Sum of Squared Errors (SSE) curve appears around **k = 4**. 

2. SSE decreases significantly up to **k = 4**, after which the rate of decrease diminishes. This indicates that adding more clusters beyond 4 does not provide a substantial improvement in clustering quality.ty.


## K-Means Clustering

```python
# Fit K-Means with optimal k (Assuming k=4 from Elbow Method)
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans_labels = kmeans.fit_predict(scaled_df)
```

```python
#Visualisation
# Ensure scaled_df is converted to NumPy array for slicing
scaled_array = scaled_df.values  # Convert DataFrame to NumPy array

# K-Means Cluster Visualization
plt.figure(figsize=(8, 6))
sns.scatterplot(x=scaled_array[:, 0], y=scaled_array[:, 1], hue=kmeans_labels, palette='viridis', s=100)
plt.title('K-Means Clustering (k=4)', fontsize=14)
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(title='Clusters', loc='upper right')
plt.show()
```

## Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler

# Fit Agglomerative Clustering with optimal k (Assuming k=4)
hierarchical = AgglomerativeClustering(n_clusters=4)
hierarchical_labels = hierarchical.fit_predict(scaled_df)
```

```python
# Dendrogram 
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Create the linkage matrix
linkage_matrix = linkage(scaled_df, method='ward')

# Plot the dendrogram
plt.figure(figsize=(12, 8))
dendrogram(
    linkage_matrix,
    truncate_mode='lastp',  # Show only the last p clusters
    p=10,                  # Number of clusters to display
    leaf_rotation=90,      # Rotate labels
    leaf_font_size=12,     # Increase font size
    show_contracted=True,  # Contract collapsed branches
    color_threshold=0.7 * max(linkage_matrix[:, 2])  # Set threshold for coloring
)
plt.title('Hierarchical Clustering Dendrogram (Truncated)', fontsize=16)
plt.xlabel('Sample Index or Cluster Size', fontsize=14)
plt.ylabel('Distance', fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.savefig('dendrogram.jpeg')
plt.show()
```

## Observations from the Dendrogram

1. **Cluster Count**:
   - The dendrogram reveals **6 prominent clusters** at the current cut-off point.
   - The clusters can be analyzed further to understand the groupings.

2. **Distance Threshold**:
   - The height of the vertical lines (distance axis) indicates the similarity or dissimilarity between clusters.
   - Clusters merging at greater heights (e.g., ~160) show more dissimilarity, while those merging at lower heights (e.g., <40) are more similar.

3. **Cluster Sizes**:
   - The numbers in parentheses (e.g., `(199)`, `(166)`) indicate the size of each cluster.
   - For example:
     - Cluster 1: 199 points
     - Cluster 2: 166 points, and so on.

4. **Optimal Clustering**:
   - A suitable distance threshold can be set slightly below **100** (based on the largest vertical gap).
   - This would yield **4 major clusters**, simplifying analysis.

5. **Hierarchical Structure**:
   - Small groups at the bottom of the dendrogram merge into larger groups, forming a hierarchy.
   - This structure helps to understand the nested relationships between clusters.


## DBSCAN Clustering

```python
# Fit DBSCAN (choose suitable eps and min_samples)
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(scaled_df)
```

```python
# Plot DBSCAN clusters
plt.figure(figsize=(10, 6))
plt.scatter(scaled_array[:, 0],scaled_array[:, 1], c=dbscan_labels, cmap='cool', marker='o', edgecolor='k', s=50)
plt.title('DBSCAN Clustering')
plt.colorbar(label='Cluster')
plt.show()
```

## Gaussian Mixture Model

```python
from sklearn.mixture import GaussianMixture

# Initialize and fit the GMM model
gmm = GaussianMixture(n_components=3, random_state=42)
gmm_labels = gmm.fit_predict(scaled_df)
```

## Evaluation and Performance Metrics:

```python
# Silhouette Score for K-Means
kmeans_silhouette = silhouette_score(scaled_df, kmeans_labels)
print(f"K-Means Silhouette Score: {kmeans_silhouette}")
```

```python
# Silhouette Score for Hierarchical Clustering
hierarchical_silhouette = silhouette_score(scaled_df, hierarchical_labels)
print(f"Hierarchical Clustering Silhouette Score: {hierarchical_silhouette}")
```

```python
# Calculate silhouette score for DBSCAN, ignoring noise points 
# Filter out noise points (-1) from DBSCAN labels
dbscan_labels_filtered = dbscan_labels[dbscan_labels != -1]
scaled_data_filtered = scaled_data[dbscan_labels != -1]

# Only calculate silhouette score if there are clusters besides noise
if len(set(dbscan_labels_filtered)) > 1:
    dbscan_silhouette = silhouette_score(scaled_data_filtered, dbscan_labels_filtered)
    print(f"Silhouette Score for DBSCAN (excluding noise): {dbscan_silhouette:.3f}")
else:
    print("DBSCAN did not find sufficient clusters for silhouette score calculation.")
```

```python
dbscan_score = silhouette_score(scaled_df, dbscan_labels)
print("\nDBSCAN Silhouette Score:", dbscan_score)
```

```python
# Calculate Silhouette Score for GMM
gmm_silhouette = silhouette_score(scaled_df, gmm_labels)
print(f"Silhouette Score for Gaussian Mixture Model: {gmm_silhouette}")
```

```python
# Create a DataFrame to store the silhouette scores
silhouette_scores = pd.DataFrame({
    "Model": ["K-Means", "Hierarchical Clustering", "DBSCAN(without noise)", "DBSCAN(with noise)","Gaussian Mixture Model"],
    "Silhouette Score": [
        kmeans_silhouette,
        hierarchical_silhouette,
        dbscan_silhouette if 'dbscan_silhouette' in locals() else "N/A",
        dbscan_score,
        gmm_silhouette
    ]
})

# Display the scores
silhouette_scores
```

## Key Observations on Clustering Models

1. **K-Means Clustering**:
   - Achieved a Silhouette Score of **0.2486**.
   - Moderate performance; further optimization (e.g., selecting the optimal number of clusters) could improve results.

2. **Hierarchical Clustering**:
   - Recorded a Silhouette Score of **0.2016**.
   - Performance is slightly lower compared to K-Means, possibly due to limitations in handling dataset complexity or the chosen linkage method.

3. **DBSCAN**:
   - **Without Noise**: Achieved the highest Silhouette Score of **0.8039**, indicating the clusters are well-separated and densely packed.
   - **With Noise**: Scored **-0.3725**, showing that the inclusion of noise points significantly impacts cluster quality.

4. **Gaussian Mixture Model (GMM)**:
   - Obtained a Silhouette Score of **0.1479**.
   - Performance was the lowest among all models, suggesting that the assumptions of GMM may not align well with the dataset's characteristics.

### Conclusion:
- **DBSCAN (without noise)** performed the best with a Silhouette Score of **0.8039**, making it the most suitable clustering model for this dataset.
- Inclusion of noise in DBSCAN and the low performance of Gaussian Mixture Model indicate the need for careful handling of outliers and alignment of model assumptions with the dataset structure.
